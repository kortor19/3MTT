```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


#load dataset
df = pd.read_csv(r"C:\Users\Bernard\Desktop\3MTT\projects\dataset\sales_dataset.csv")
df
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
#set seaborn style
sns.set_theme(style="whitegrid")

#Distribution plot for numerical data
for col in df.select_dtypes(include=['object']).columns:
    plt.figure(figsize=(10,4))
    sns.countplot(data=df, x=col)
    plt.title(f'Count of {col}')
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Correlation heatmap (for numerical data). filter specific numerical columns
if len(df.select_dtypes(include=['float64', 'int64']).columns) > 1:
    numerical_cols = ['Quantity', 'Price']
    df_filtered = df[numerical_cols]
    plt.figure(figsize=(10,8))
    sns.heatmap(df_filtered.corr(), annot=True, cmap='coolwarm')
    plt.title('Correlation Heatmap')
    plt.show()
```

```python
# Pairplot for numerical columns (for smaller datasets)
if len(df.select_dtypes(include=['float64', 'int64']).columns) <= 5:
    sns.pairplot(df)
    plt.show()
```
